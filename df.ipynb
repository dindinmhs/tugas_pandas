```python
import pandas as pd
import numpy as np

raw = pd.read_csv('dataset.csv')

raw
```

```python
df = raw.iloc[:,4:8]
df
```

```python
jumlah_per_tahun = {
    'tahun' : [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'total_produksi' : [0] * 9,
    'satuan' : ['TON PER HARI'] * 9
}

for idx in range(len(jumlah_per_tahun['tahun'])) :
    for i, row in df.iterrows() :
        if row['tahun'] == jumlah_per_tahun['tahun'][idx] :
            jumlah_per_tahun['total_produksi'][idx] += row['jumlah_produksi_sampah']
            
df2 = pd.DataFrame(jumlah_per_tahun)
df2
```

```python
df
df_cleaned = df.dropna()
df_cleaned
```

```python
jumlah_per_kota = {
    'nama_kab_kota' : [],
    'total_produksi' : [0],
    'satuan' : ['TON PER HARI']
}

for i, row in df_cleaned.iterrows() :
    if row['nama_kabupaten_kota'] not in jumlah_per_kota['nama_kab_kota'] :
        jumlah_per_kota['nama_kab_kota'].append(row['nama_kabupaten_kota'])
        
panjang = len(jumlah_per_kota['nama_kab_kota'])

jumlah_per_kota['total_produksi'] *= panjang
jumlah_per_kota['satuan'] *= panjang

for idx in range(panjang) :
    for i, row in df_cleaned.iterrows() :
        if row['nama_kabupaten_kota'] == jumlah_per_kota['nama_kab_kota'][idx] :
            jumlah_per_kota['total_produksi'][idx] += row['jumlah_produksi_sampah']

df3 = pd.DataFrame(jumlah_per_kota)
df3
```

```python
df2.to_csv('jumlah_per_tahun.csv')
df3.to_csv('jumlah_per_kota.csv')
```
